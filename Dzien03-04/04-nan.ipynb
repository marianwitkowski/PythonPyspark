```
from pyspark.sql.functions import isnan, when, count, col

df.select([ count( when(isnan(c) | col(c).isNull(), c )).alias(c) for c in df.columns]).show()
```

```
from pyspark.sql.functions import isnan, isnull

#display( df.filter( df.Name.isNull() ) )
display( df.filter( isnan(df.Name) | isnull(df.Name) ) )
```

```
# usuwanie wierszy, gdzie we wszystkich kolumnach s NaN/null
df = df.dropna(how="all", subset=df.columns)
```

```
df_tmp = df.fillna(12.34, subset="Salary")
display(df_tmp)
```
