```
url = "https://stooq.pl/q/d/l/?s=pkn&d1=20210101&d2=20230620&i=d"
df = pd.read_csv(url)
df
```

```
df.dtypes
```

```
df.Data = pd.to_datetime(df.Data, format="%Y-%m-%d")
df.dtypes
```

```
df.set_index("Data", inplace=True)
df.head()
```

```
df.Zamkniecie.plot(grid=True, color="red", figsize=(11,6))
```

```
df.Zamkniecie.resample("14D").max().plot(grid=True, color="red", figsize=(11,6))
```

```
df.Zamkniecie.resample("MS").mean().plot(grid=True, color="red", figsize=(11,6), kind="bar")
```
